```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('train.csv')
print("Размер датасета:", df.shape)
print("Первые 5 строк:")
df.head()
```

```python
print("Информация о датасете:")
df.info()
```

```python
print("Типы данных:")
df.dtypes
```

```python
print("Названия столбцов:")
df.columns.tolist()
```

```python
print("Пропущенные значения в исходном датасете:")
print(df.isnull().sum())
```

```python
df_filled = df.copy()

df_filled['Age'] = df_filled['Age'].fillna(df_filled['Age'].median())
df_filled['HomePlanet'] = df_filled['HomePlanet'].fillna(df_filled['HomePlanet'].mode()[0])
df_filled['CryoSleep'] = df_filled['CryoSleep'].fillna(df_filled['CryoSleep'].mode()[0])
df_filled['Cabin'] = df_filled['Cabin'].fillna(df_filled['Cabin'].mode()[0])
df_filled['Destination'] = df_filled['Destination'].fillna(df_filled['Destination'].mode()[0])
df_filled['VIP'] = df_filled['VIP'].fillna(df_filled['VIP'].mode()[0])
df_filled['RoomService'] = df_filled['RoomService'].fillna(df_filled['RoomService'].median())
df_filled['FoodCourt'] = df_filled['FoodCourt'].fillna(df_filled['FoodCourt'].median())
df_filled['ShoppingMall'] = df_filled['ShoppingMall'].fillna(df_filled['ShoppingMall'].median())
df_filled['Spa'] = df_filled['Spa'].fillna(df_filled['Spa'].median())
df_filled['VRDeck'] = df_filled['VRDeck'].fillna(df_filled['VRDeck'].median())
df_filled['Name'] = df_filled['Name'].fillna(df_filled['Name'].mode()[0])

print("Проверка после заполнения пропусков:")
print(df_filled.isnull().sum())
print("Всего пропусков:", df_filled.isnull().sum().sum())
```

```python
numeric_cols = df_filled.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Числовые столбцы для нормализации:", numeric_cols)

df_normalized = df_filled.copy()
scaler = MinMaxScaler()

for col in numeric_cols:
    df_normalized[col] = scaler.fit_transform(df_normalized[[col]])

print("Первые 5 строк после нормализации (числовые столбцы):")
print(df_normalized[numeric_cols].head())
```

```python
df_normalized['Cabin_Deck'] = df_normalized['Cabin'].str[0]
df_normalized['Cabin_Side'] = df_normalized['Cabin'].str.split('/').str[2]
df_normalized['Name_FirstLetter'] = df_normalized['Name'].str[0]

df_normalized = df_normalized.drop(['Name', 'Cabin','PassengerId'], axis=1)

categorical_cols_to_encode = df_normalized.select_dtypes(include=['object', 'string']).columns.tolist()
df_encoded = pd.get_dummies(df_normalized, columns=categorical_cols_to_encode)

print("Размер датасета после кодирования:", df_encoded.shape)
print("Первые 5 строк после категоризации:")
print(df_encoded.head())
```

```python
df_encoded.to_csv("rezult_titanic.csv", index=False)
```
